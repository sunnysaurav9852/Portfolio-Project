```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read in the data
df = pd.read_csv(r'C:\Users\Sunny Saurav\Downloads\movies.csv')
```

```python
df.head()
```

```python
#finding missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.fillna(0,inplace = True)
```

```python
#finding missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Datatypes of each columns

df.dtypes
```

```python
#change datatypes of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#correct year

df['yearcorrect'] = df['year'].astype(str).str[0:4]
```

```python
df
```

```python
df.sort_values(by=['gross'], inplace = False , ascending = False)
```

```python
#drop any duplicates
df.drop_duplicates()
```

```python
df.sort_values(by='budget',inplace = False,ascending = False)
```

```python
#scatter plot betweeen budget and gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget of the Film')
plt.ylabel('Gross Earnings of the Film')
plt.show()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"blue"},line_kws={"color":"red"})
```

```python
#let's start looking at correlation

numerical_df = df.select_dtypes(include = ['number'])
correlation_matrix = numerical_df.corr(method = 'pearson') #pearson, kendall, spearman
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtypes == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
correlation_matrix2 = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix2,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
